```python
from pyecharts.charts import Map
from pyecharts import options as opts
import pandas as pd

province_data = pd.read_excel(r'/Users/lye/Documents/图表小挑战/武汉冠状病毒疫情/各省疫情数据2020-02-05.xlsx',usecols=[2,3])
# province_data = df.to_dict(orient='dict')

province_data.confirmedCount = (province_data['confirmedCount']).astype(str)
province_data.provinceShortName = (province_data['provinceShortName']).astype(str)
province_dict = province_data.set_index('provinceShortName')['confirmedCount'].to_dict()
# print(province_dict)

map_CN = Map()
map_CN.set_global_opts(
    title_opts=opts.TitleOpts(title='20200205中国疫情地图'),
    visualmap_opts=opts.VisualMapOpts(max_=10000,is_piecewise=True,
                                     pieces =[
                                         {"max": 50000, "min": 1001, "label": ">1000", "color": "#8A0808"},
                                         {"max": 1000, "min": 500, "label": "500-1000", "color": "#B40404"},
                                         {"max": 499, "min": 100, "label": "100-499", "color": "#DF0101"},
                                         {"max": 99, "min": 10, "label": "10-99", "color": "#F78181"},
                                         {"max": 9, "min": 1, "label": "1-9", "color": "#F5A9A9"},
                                         {"max": 0, "min": 0, "label": "0", "color": "#FFFFFF"},
                                     ],)
)

map_CN.add('20200205中国疫情地图',data_pair=province_dict.items(),maptype='china',is_roam=True)
map_CN.render_notebook()
```
